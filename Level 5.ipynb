```python
class Triangle:
    def __init__(self,s1,s2,s3):
        self.side1 = s1
        self.side2 = s2
        self.side3 = s3
        
    def area_(self):
        s = (self.side1+self.side2+self.side3)/2
        area = (s*(s-self.side1)*(s-self.side2)*(s-self.side3))**0.5
        area = float(area)
        return area
    
    def perimeter_(self):
        p = self.side1+self.side2+self.side3
        return p
    def tell_type(self):
        if( self.side1 == self.side2 ==self.side3 ):
            print("The Given triangle is an equilateral triangle.")
        elif( self.side1 == self.side2 or self.side1 ==self.side3 ):
            print("The Given triangle is an isosceles triangle.")
        else:
            print("The Given triangle is an scalene triangle.")
            
s_1 = int(input("Enter 1st side: "))
s_2 = int(input("Enter 2nd side: "))
s_3 = int(input("Enter 3rd side: "))
triangle = Triangle(s_1,s_2,s_3)
print("Side 1 is: ",triangle.side1)
print("Side 2 is: ",triangle.side2)
print("Side 3 is: ",triangle.side3)
print("Area is: ",triangle.area_())
print("Perimeter is: ",triangle.perimeter_())
triangle.tell_type()
```

```python
class Pyramid:
    def __init__(self):
        pass
    def show_(self,n,op):
        for i in range(n):
            for j in range(i+1):
                print(op, end = " ")  
            print("\r")
            
    def show_odd(self,n,op):
        c = -1
        for i in range(n):
            c+=1
            for j in range(i+c+1):
                print(op,end = " ") 
            print("\r")
            
    def show_reverse(self,n,op):
        for i in range(n):
            for j in range(n-i):
                print(op, end = " ")  
            print("\r")
            
    def show_1(self,n,op):  
        for i in range(n):
            for j in range(i+1):
                print(op, end = " ")  
            print("\r")
        for i in range(n):
            for j in range(n-i):
                print(op, end = " ")  
            print("\r")
    
p = Pyramid()
n = int(input("Enter size: ")) 
o = input("Enter operator: ")
p.show_(n,o)
n = int(input("Enter size: "))
o = input("Enter operator: ")
p.show_odd(n,o)
n = int(input("Enter size: "))
o = input("Enter operator: ")
p.show_reverse(n,o)
n = int(input("Enter size: "))
o = input("Enter operator: ")
p.show_1(n,o)
```

```python
import time
class Cake:
    def __init__(self,flour,egg,sugar,milk,baking_soda):
        self.Flour = flour
        self.Egg = egg
        self.Sugar = sugar
        self.Milk = milk
        self.Temperature = 50
        self.Baking_soda = baking_soda
        self.Batter_mixed = False
        self.foamy_egg_whites = False
        
    def beat_eggs_whites(self):
        for i in range(5):
            print("Beating egg whites.")
            time.sleep(1)
        print("Egg-whites are foamy now.")
        self.foamy_egg_whites = True
        
    def mix(self):
        self.Batter_mixed = True
        print("Ingredients are mixed.Ready to bake.")
        
    def bake(self):
        if self.Batter_mixed and self.foamy_egg_whites:
            while self.Temperature<250:
                print(self.Temperature)
                self.Temperature+=20
                time.sleep(1)
        print("Your cake is Baked now.Temprature : ", self.Temperature)
        
        
class CreanmCake(Cake): 
    def __init__(self,flour,egg,sugar,milk,baking_soda,cream):
        super(CreanmCake,self).__init__(flour,egg,sugar,milk,baking_soda)
        self.Cream = cream
        
    def add_cream(self):
        self.Cream = True
        print("Now we will add Cream.\nCream Added.")
        print("If you want to add toppings on cake.First refrigerate it.")
    def refrigerate(self):
        while self.Temperature>0:
            print(self.Temperature)
            self.Temperature-=10
            time.sleep(1)
        print("Take out the cake from the Refrigerator.Temprature : ", self.Temperature, "\n\n")

class CherryCreamCake(CreanmCake):
    def __init__(self,flour,egg,sugar,milk,baking_soda,cream,cherry):
        super(CherryCreamCake,self).__init__(flour,egg,sugar,milk,baking_soda,cream)
        self.Cherry = cherry
        
    def add_cherry(self):
        self.Cherry = True
        time.sleep(1)
        print("Cherry added.")
class ChocolateCherryCreamCake(CherryCreamCake): 
    def __init__(self,flour,egg,sugar,milk,baking_soda,cream,cherry,chocolate):
        super(ChocolateCherryCreamCake,self).__init__(flour,egg,sugar,milk,baking_soda,cream,cherry)
        self.Chocolate = chocolate
        
    def add_chocolate(self):
        self.Chocolate = True
        time.sleep(1)
        print("Chocolate added.")
    
    def Cake_ready(self):
        if self.Cherry and self.Chocolate and self.Cream:
            time.sleep(1)
            print("Your Cake is ready to serve now.")
        else:
            time.sleep(1)
            print("Your cake is not ready yet. Just a few more steps.")
```
